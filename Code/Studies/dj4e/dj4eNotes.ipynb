remember the guide to making model changes:

 - Change your models (in models.py).
 - Run python manage.py check
 - Run python manage.py makemigrations to create migrations for those changes
 - Run python manage.py migrate to apply those changes to the database.

The reason that there are separate commands to make and apply migrations is because youll commit migrations to your version control system and ship them with your app; they not only make your development easier, theyre also usable by other developers and in production.
